```python
import pandas as pd
df = pd.read_csv("data/student_marks.csv")
df
```

```python
# Total and Average
df["Total"] = df["English"] + df["Maths"] + df["Science"] + df["History"]
df["Average"] = df["Total"] / 4

df
```

```python
# Highest & Lowest scorer
topper = df.loc[df["Total"].idxmax()]
lowest = df.loc[df["Total"].idxmin()]
print("Topper:")
print(topper)
print("\nLowest Scorer:")
print(lowest)
```

```python
grades = []
for avg in df["Average"]:
    if avg >= 80:
        grades.append("A")
    elif avg >= 65:
        grades.append("B")
    elif avg >= 45:
        grades.append("C")
    else:
        grades.append("Fail")
df["Grade"] = grades
df
```

```python
import matplotlib.pyplot as plt
plt.bar(df["Name"], df["Total"])
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.title("Student Marks Analysis")
plt.show()
```
